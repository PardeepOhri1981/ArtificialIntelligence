```python
###IRIS Data Set ## To be Learn based on Sepal and Petal length and width we are categorising the Species
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snc
import warnings 
warnings.filterwarnings('ignore')

iris = pd.read_csv('IRIS.csv') ## IRIS dataset file Iric.csv  (Downloaded form Kaggle)
print("dataSet Shape \n",iris.shape )## provide the shape of iris
print("DataSet Columns: \n",iris.columns)
print("Check the Species category",iris['species'].value_counts()) ## Dataset should not be biased to any category
for col in iris.columns:
    print('\n',col)
    print(iris[col].unique())
```

```python
 ## CDF  - Commulative Distributoon function
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=  10, density=True)
pdf = counts /(sum(counts)) ## Formula to create bins
print(pdf)
print(bin_edges)
```

```python
cdf = np.cumsum(pdf)
plt.xlabel("Setosa-Petal Legth")
plt.ylabel("Setosa-Petal Legth")

plt.plot(bin_edges[1:],pdf,label = 'pdf')
plt.plot(bin_edges[1:],cdf, label = 'cdf')
plt.legend()
```

```python
counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=  10, density=True)
pdf = counts /(sum(counts)) ## Formula to create bins
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.xlabel("Versicolor-Petal Legth")
plt.ylabel("Versicolor-Petal Legth")

plt.plot(bin_edges[1:],pdf,label = 'pdf')
plt.plot(bin_edges[1:],cdf, label = 'cdf')
plt.legend()
```

```python
 ## CDF  - Commulative Distributoon function
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=  10, density=True)
pdf = counts /(sum(counts)) ## Formula to create bins
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.xlabel("Setosa-Petal Legth")
plt.ylabel("Setosa-Petal Legth")

plt.plot(bin_edges[1:],pdf,label = 'Setosa pdf')
plt.plot(bin_edges[1:],cdf, label = ' Setosa cdf')
plt.legend()


counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=  10, density=True)
pdf = counts /(sum(counts)) ## Formula to create bins
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.xlabel("Petal Legth")
plt.ylabel("Probablity")

plt.plot(bin_edges[1:],pdf,label = 'Versicolor pdf')
plt.plot(bin_edges[1:],cdf, label = 'Versicolor cdf')
plt.legend()

counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins=  10, density=True)
pdf = counts /(sum(counts)) ## Formula to create bins
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.xlabel("Petal Legth")
plt.ylabel("Probablity")

plt.plot(bin_edges[1:],pdf,label = 'Virginca pdf')
plt.plot(bin_edges[1:],cdf, label = 'Virginca cdf')
plt.legend()
```

```python
## Mean Variance and Standard Deviation

print("\n Mean")
print('Seatosa Petal Length :: ',np.mean(iris_setosa['petal_length']))
print('Versicolor Petal Length ::',np.mean(iris_versicolor['petal_length']))
print('Viginica Petal Length :: ',np.mean(iris_virginica['petal_length']))
```

```python
print("\n Mean with outlier") ## aading one higher or lower value to check the impact the mean, which mean means is highly impacted due to adding one higher or lower value, mean not recommended
print('Seatosa Petal Length :: ',np.mean(np.append(iris_setosa['petal_length'],50)))
print('Versicolor Petal Length ::',np.mean(np.append(iris_versicolor['petal_length'],50)))
print('Viginica Petal Length :: ',np.mean(np.append(iris_virginica['petal_length'],50)))
```

### Standard Deviation

```python
## standard deviation is a measure of the amount of variation of a random variable expected about its mean. but its not accurate as its missed the bounderies of spread
print('Spread of Seatosa petal length',np.std(iris_setosa['petal_length']))
print('Spread of Versicolor petal length',np.std(iris_versicolor['petal_length']))
print('Spread of Virginica petal length',np.std(iris_virginica['petal_length']))
```

#Median

```python
print('Median')
print('Spread of Seatosa petal length',np.median(iris_setosa['petal_length']))
print('Spread of Versicolor petal length',np.median(iris_versicolor['petal_length']))
print('Spread of Virginica petal length',np.median(iris_virginica['petal_length']))
```

```python
## up to maximum (50% of normal value count of column ) Outliers can be added to get the accurate median

print('Median with outlier') ## To check the accuracy
print('Spread of Seatosa petal length',np.median(np.append(iris_setosa['petal_length'],500)))
print('Spread of Versicolor petal length',np.median(np.append(iris_versicolor['petal_length'],500)))
print('Spread of Virginica petal length',np.median(np.append(iris_virginica['petal_length'],500)))
```

## Percentile

```python
print(' \n Calculation 90 Percentile') ## We can change the percentile
print('Spread of Seatosa petal length',np.percentile(iris_setosa['petal_length'],90))
print('Spread of Versicolor petal length',np.percentile(iris_versicolor['petal_length'],90))
print('Spread of Virginica petal length',np.percentile(iris_virginica['petal_length'],90))
```

## Quantile

```python
print(' \n Calculation Quantiles') ## We can change the Quantile
print('Spread of Seatosa petal length',np.percentile(iris_setosa['petal_length'],np.arange(0,100,25)))
print('Spread of Versicolor petal length',np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25)))
print('Spread of Virginica petal length',np.percentile(iris_virginica['petal_length'],np.arange(0,100,25)))
```
