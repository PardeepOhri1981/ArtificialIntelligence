```python
import matplotlib.pyplot as plt
plt.plot([2,4,6,4])
plt.ylabel("Number")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.show()
```

```python
plt.plot([1,4,2,4],[2,4,6,6])
plt.ylabel("Number")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.grid()
plt.gray()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,6,8],'r^')
plt.ylabel("Number")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.grid()
plt.gray()
plt.show()
```

```python
import numpy as np
t = np.arange(0.0, 5.0,0.2)
plt.plot(t,t**2,'b--', label = '^2')
plt.plot(t,t**2.2,'ro', label = '^2.2')
plt.plot(t,t**2.5,'g^', label = '^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python

x = [1,2,3,4]
y = [1,4,6,16]
plt.plot(x,y,linewidth = 2,label = '^2')
plt.grid()
plt.legend()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color = 'r', linewidth = 2.0, label='Red Line')
plt.setp(lines[1],color = 'g', linewidth = 2.0, label='Green Line')
plt.grid()
plt.legend()
plt.show()
```

```python
import math
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0, 5.0,0.02)
plt.figure(1)
plt.subplot(211) ## Creating 2 subplot and 1 Figure  1 graph
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212) ## Creating 2 subplot and 1 subplot w graph
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r^')

plt.show()
```

```python
import math
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0, 5.0,0.02)
t3 = np.arange(0.0, 5.0,0.03)
plt.figure(1)
plt.subplot(321) ## Creating 3 rows, 2 column first draw
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(323) ## Creating 3 subplot and 1 subplot 2 graph
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r^')

plt.subplot(322) ## Creating 3 subplot and 1 subplot 2 graph
plt.grid()
plt.plot(t3,np.sin(2*np.pi*t3),'g^')

plt.show()
```
