```python
## Merging and Concate

india_Weather = pd.DataFrame({'City':['Mumbai','Delhi', 'Chandigarh'], 'Temperature': [32, 35,30], 'Humidity': [68,65,79]})
print("India Weather: \n",india_Weather)
us_Weather = pd.DataFrame({'City':['New york','Chicago', 'Orland'], 'Temperature': [21, 14,23], 'Humidity': [68,65,75]})
print("US Weather: \n",us_Weather)

print("Concatination with index:\n",pd.concat([india_Weather, us_Weather]))
print("Concatination without index:\n",pd.concat([india_Weather, us_Weather],ignore_index= True))
```

```python
print("Concatination with axis is 1:\n",pd.concat([india_Weather, us_Weather],axis= 1))
print("Concatination with axis 0:\n",pd.concat([india_Weather, us_Weather],axis=0))
```

```python
print("Merging with inner:\n",pd.merge(india_Weather, us_Weather, on='Humidity',how='inner'))
# print("Concatination with outer\n",pd.concat([india_Weather, us_Weather],axis=0))
```

```python
print("Merging with inner:\n",pd.merge(india_Weather, us_Weather, on='Humidity',how='outer'))
```

```python
print("Merging with inner:\n",pd.merge(india_Weather, us_Weather, on='Humidity',how='left'))
```

```python
print("Merging with inner:\n",pd.merge(india_Weather, us_Weather, on='Humidity',how='right'))
```

```python
print("Merging with inner:\n",pd.merge(india_Weather, us_Weather,how='cross'))
```
