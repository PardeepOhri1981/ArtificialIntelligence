```python
habermanDataFrame_Survival= habermanDataFrame.loc[habermanDataFrame['SurvivalStatus'] == 1]
habermanDataFrame_NotSurvival= habermanDataFrame.loc[habermanDataFrame['SurvivalStatus'] == 2]
print(habermanDataFrame_Survival)
print(habermanDataFrame_NotSurvival)
```

```python
# 1D graph for accuracy of Survival Status

## PDF - Probablity Desnsity function

plt.plot(habermanDataFrame_Survival['Nodes'], np.zeros_like(habermanDataFrame_Survival['Nodes']))
plt.plot(habermanDataFrame_NotSurvival['Nodes'], np.zeros_like(habermanDataFrame_NotSurvival['Nodes']))
plt.xlabel('SurvivalStatus')
plt.ylabel('Nothing')
plt.show()
```

```python
## PDF - Probablity Desnsity function

plt.plot(habermanDataFrame_Survival['OperationYear'], np.zeros_like(habermanDataFrame_Survival['OperationYear']))
plt.plot(habermanDataFrame_NotSurvival['OperationYear'], np.zeros_like(habermanDataFrame_NotSurvival['OperationYear']))
plt.xlabel('SurvivalStatus')
plt.ylabel('Nothing')
plt.show()
```

```python
## PDF - Probablity Desnsity function

plt.plot(habermanDataFrame_Survival['Age'], np.zeros_like(habermanDataFrame_Survival['Age']))
plt.plot(habermanDataFrame_NotSurvival['Age'], np.zeros_like(habermanDataFrame_NotSurvival['Age']))
plt.xlabel('SurvivalStatus')
plt.ylabel('Nothing')
plt.show()
```
