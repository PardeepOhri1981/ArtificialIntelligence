```python
import matplotlib.pyplot as plt
## Draw graph with one axis only
plt.plot([2,4,6,8],'^--b',)
plt.ylabel('Number')
plt.xlabel('Indices')
plt.title('My Plot')
plt.grid()
plt.gray()
plt.show()
```

```python
## Draw graph with 2 Axis
plt.plot([2,4,6,8],[8,6,4,2],'^--b')
plt.ylabel('Number')
plt.xlabel('Indices')
plt.title('My Plot')
plt.box(True) ### Graphs in boxes
# plt.arrow(2,4,6,8) ## Draw arroe in the graph
plt.autoscale(True)  ## Draw the graphs
# plt.axes()
plt.gray()
plt.show()
```

```python
import numpy as np
t = np.arange(0.0, 5.0,0.2)
plt.plot(t,t**2,'b--', label = '^2')
plt.plot(t,t**2.2,'ro', label = '^2.2')
plt.plot(t,t**2.5,'g^', label = '^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [2,10,6,8]
plt.plot(x,y,linewidth = 6,label = '^2')
plt.grid()
plt.legend()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],  color='r', linewidth= 6,label ='Red Line' )
plt.setp(lines[1], color='g', linewidth= 4,label ='Green Line' )
plt.gray()
plt.grid()
plt.show()
```

```python
import math
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0, 5.0,0.02)
plt.figure(1)
plt.subplot(211) ## Creating 2 subplot and 1 Figure  1 graph
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212) ## Creating 2 subplot and 1 subplot w graph
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r^')

plt.figure(2)
plt.subplot(221) ## Creating 2 subplot and 1 Figure  1 graph
plt.grid()
plt.plot(t1,t1**3,'b--')

plt.subplot(222) ## Creating 2 subplot and 1 subplot w graph
plt.grid()
plt.plot(t2,t2**2,'r^')

plt.show()
```
