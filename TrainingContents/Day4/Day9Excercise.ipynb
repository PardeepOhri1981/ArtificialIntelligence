IRIs Data Set get from git hb

```python
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
print(df)
df.plot(figsize=(20,8))
```

```python
df.plot(kind = 'bar', figsize=(20,8))
```

```python
df.iloc[5:15].plot(kind = 'bar',figsize=(20,8))
df.iloc[20:50].plot(kind = 'bar',figsize=(20,8))
df.iloc[[149]].plot(kind = 'bar',figsize=(20,8))
df.iloc[20:30].plot(kind = 'barh',figsize=(20,8))
```

```python
df.plot(kind='hist',figsize=(20,8))
df.plot(kind='hist')
df.plot(kind='hist',alpha = .8) ## to make transparent
```

```python
df['sepal_length'].plot(kind='hist',alpha = .5)
```

```python
df[['sepal_length','sepal_width', 'petal_length']].plot(kind='hist',alpha = .5,orientation='horizontal')
```

```python
df.hist(figsize=(20,8),alpha = .6,color='r')
```

```python
#Box Plot
df.plot(kind='box',figsize=(20,8))
```

```python
print(df.describe()) 
## Min , max value od box plor Max value (Q3+1.5*IQR), Min value  =(Q1- 1.5*IQR)  and IQR(inter Quartile range) = (75%-25%), Q3 = 3 quarter i.e 75% q1 = 1 Quarter i.e 25%
df.plot(kind='box',figsize=(20,8),color={'boxes':'g','whiskers':'r'},vert=False)
```

```python
df.plot(kind='area',figsize=(20,8),alpha = .9)
```

```python
df.plot(kind='area',figsize=(20,8),alpha = .9, stacked= False)
```

```python
df.plot.scatter(x='sepal_length',y='petal_length') ## 2d grpah
df
```

```python
df.plot.scatter(x='sepal_length',y='petal_length', c= 'petal_width') # 3 dimesions
```

```python
df.plot.scatter(x='sepal_length',y='petal_length', c= 'petal_width', s=df['sepal_width']*10) ## 4 dimension scatter graph x,y,c,d 4 dimension
df
```

```python
df.plot.hexbin(x='sepal_length',y='petal_length', C= 'petal_width')
```

```python
##Plotly
df1 = df.drop('species',axis=1).iloc[0]
df1.plot(kind='pie')
```

```python
df2 = pd.DataFrame(df.tail(1))  ## Get last row and draw pie
df2.reset_index(inplace=True,drop=True)
df3 = df2.drop('species',axis=1).iloc[0]
df3.plot(kind='pie')
```
