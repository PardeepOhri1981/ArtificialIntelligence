```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import  load_iris
```

```python
iris = load_iris()
iris
```

```python
X= iris.data
y= iris.target
```

```python
X
```

```python
y
```

```python
#Binary Output
from sklearn.preprocessing import label_binarize
y_binary = label_binarize(y,classes=[0,1,2])
y_binary
```

```python
n_classes = y_binary.shape[1]
n_classes
```

```python
# Train - Test split
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y_binary,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
## Random Forest
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
rf = OneVsRestClassifier(RandomForestClassifier(n_estimators=100, random_state=42))
```

```python
rf
```

```python
rf.fit(X_train,y_train)
```

```python
## Probability score or probability of y predict for ROC curve
y_predict_prob = rf.predict_proba(X_test)
y_predict_prob
```

```python
## Compute AUC score
from sklearn.metrics import roc_auc_score, roc_curve, auc
auc_score = []
plt.figure(figsize=(8,6))
colors = ['red', 'blue', 'green']
for i in range(n_classes):
    fpr, tpr,_ = roc_curve(y_test[:,i], y_predict_prob[:,i])
    roc_auc = auc(fpr,tpr)
    print(f"Class{iris.target_names[i]} : AUc = {roc_auc: .3f}")
   # print(f'clas{iris.target_names[i]}  - FPR : {fpr})
   # print(f'class{iris.target_names[i]} - TPR : {tpr})
    plt.plot(fpr,tpr,color = colors[i],label = f'Class{iris.target_names[i]}(AUC = {roc_auc:.3f})')
plt.plot([0,1],[0,1],'k--')

```
