```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5.0, 5.0,100)
x = np.hstack([x, -x])
x
```

```python
y = np.sqrt(10**2 - x**2)
y = np.hstack([y, -y])
y
```

```python
x1 = np.linspace(-5.0, 5.0,100)
y1 =np.sqrt(5**2 - x1**2)
y1 = np.hstack([y1, -y1])
x1= np.hstack([x1, -x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
np.vstack([y,x])
```

```python
import pandas as pd
df1 = pd.DataFrame(np.vstack([y,x]).T,columns =['X1','X2'])
df1['Y'] = 0
df1
```

```python
df2 = pd.DataFrame(np.vstack([y1,x1]).T,columns =['X1','X2'])
df2['Y'] = 1
df2
```

```python
df = pd.concat([df1, df2],ignore_index=True)
df.head(20)
```

```python
df.shape
X = df.iloc[:,:2]
y = df.Y
```

```python
from sklearn.model_selection import train_test_split

X_Train, X_Test, Y_train, y_Test = train_test_split(X, y, test_size=0.25, random_state= 42)
Y_train
```

```python
y_Test
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(X_Train, Y_train)
```

```python
from sklearn import metrics
y_predict = classifier.predict(X_Test)
accuracy = metrics.accuracy_score(y_Test,y_predict)
accuracy
```

## Polynomial - Convert 1D to muliple D

## Support  Vector Classifier with Marginal Value

```python
from scipy import stats
import seaborn as sns
from sklearn.datasets import make_blobs
X,Y = make_blobs(n_samples=100, centers=2,random_state=0,cluster_std=0.6)# centers will define the value in 2nd parameters i.e y, Cluster_std: more value overlap, less value classified
plt.scatter(X[:,0],X[:,1],c=Y,s=50,cmap='winter')   
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear',C=1)
model.fit(X,Y)
```

```python
def plot_svc_decision_function(model, ax = None,plot_support = True):
    if ax is None:
        ax = plt.gca() # get current axis
        xlim = ax.get_xlim()
        ylim=ax.get_ylim()
        x = np.linspace(xlim[0],xlim[1],30)
        y = np.linspace(ylim[0],ylim[1],30)
        Y,X = np.meshgrid(y,x)

        xy=np.vstack([X.ravel(),Y.ravel()]).T
        P= model.decision_function(xy).reshape(X.shape)
        ax.contour(X,Y,P,color = 'k',
                   levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])
        if plot_support:
            ax.scatter(model.support_vectors_[:,0],
                       model.support_vectors_[:,1],
                       s=300,linewidth=1,facecolors='none')
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=50,cmap='winter')
plot_svc_decision_function(model)       
```
