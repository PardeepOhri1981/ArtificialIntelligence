```python
x= dataset[["TV"]]
x
```

```python
y = dataset[["Sales"]]
y
```

```python
# sklearn  model used  in Machine Learning for Linear Regression
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
#Data set is small so not using Train Test model
linear_model.fit(x,y)
```

```python
linear_model.intercept_    ### In line equation y = mx+c the value of c is interceptor or constant
```

```python
linear_model.coef_     ## In line equation y =mx+c, the value of m is coeffecient or slope
```

```python
import pickle ## to Save the Model using pickle
file = "linear_reg.sav"  # to save the regression  model file
pickle.dump(linear_model, open(file,"wb"))  
```

```python
# to check the investment  45 in TV impact on Sales
linear_model.predict([[45]]) ## Calculation of y prediction.
```

```python
list = [5,7,9,12,45,67,89,100]
for i in list:
    print(linear_model.predict([[i]]))
```

```python
## To calculate R2 for accuracy of cost function -
linear_model.score(x,y)  ## Accuracy of cost function  or Error Rate or Best Fit line is 61.18%
```

```python
### We can also use the state model in place of linear model
import statsmodels.formula.api as smf
lm = smf.ols(formula='Sales ~ TV', data=dataset).fit()   ## ols means oridinary least square
lm.summary() ## Check the probablity value
```

```python
### We can also use the state model in place of liner model
import statsmodels.formula.api as smf
lm = smf.ols(formula='Sales ~ TV+Radio', data=dataset).fit()   ## ols means oridinary least square of TV and Radio
lm.summary()

# here probability value id 0 whih means its 100 % impacted the Sales price
```

```python
### We can also use the state model in place of liner model
import statsmodels.formula.api as smf
lm = smf.ols(formula='Sales ~ TV+Radio+Newspaper', data=dataset).fit()   ## ols means oridinary least square of TV, Newspaper and Radio
lm.summary()

# here probability value  of newspaper is not 0 whih means its not impacted the Sales price we can exclude it
```
