# Bimary NB Classifier


```python
df = pd.read_csv('PlayTennis.csv')
df
```

```python
## Label Encoder
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
    
```

```python
print(label_encoders)
print(df)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X = df.drop('Play Tennis', axis =1)
y = df['Play Tennis']
print(X)
print(y)
```

```python
X_train, X_Test, y_train, y_test = train_test_split(X,y,test_size= 0.4, random_state=42) # size is % of data

print(X_train)
print(y_train)
```

```python
# Naive Baise Classifier
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
##prediction
y_prediction = model.predict(X_Test)
```

```python
#model Evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
confusionmatrix = confusion_matrix(y_test, y_prediction)
```

```python
print('confusion_matrix :', confusionmatrix)
```

```python
classificationreport = classification_report(y_test, y_prediction)
print('ClassificationReport :', classificationreport)
```

```python
print('Accuracy:', accuracy_score(y_test, y_prediction))
```
