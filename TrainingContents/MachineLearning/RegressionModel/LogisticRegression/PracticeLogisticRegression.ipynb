## Logistic Regression with Regularization  https://archive.ics.uci.edu/dataset/45/heart+disease


```python
df = pd.read_csv('processed.cleveland.data')
df
columns = ["age","sex", "cp","trestbos","chol","fbs", "restecg","thalach","exand","oldpeach", "slop", "cs","thal","target"]
data = pd.read_csv('processed.cleveland.data', names=columns, header= None, na_values= "?")
data = data.reset_index(drop = True)
data
```

```python
data = data.dropna()
data
```

```python
# convert into binary classification
data['target']= (data['target'] > 0).astype(int)
data
```

```python
X = data.drop('target', axis = 1)
y = data['target']
X_no_header = pd.DataFrame(X.values)
Y_no_header = pd.Series(y.values)
print(X_no_header)
```

```python
## Normalise Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_no_header)
X_scaled
```

```python
#Train Test Split
X_train, X_test,y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state= 42)
X_train
y_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score (y_test, y_pred)
confusion_matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(accuracy)
```

```python
print(report)
```

#Logistic Regression with L1 and L2 Regularizer

```python
# L1 Regularlizer - (Lasso)
model_l1 = LogisticRegression(penalty ='l1', solver = 'liblinear', C = 1.0) ## Liblinear id use for identify and ignore the less important feature
```

```python
model_l1.fit(X_train, y_train)
predict_l1_y_test = model_l1.predict(X_test)
predict_l1_y_test
print(y_test.shape)
print(predict_l1_y_test.shape)
```

```python
#Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
confusion_matrix_l1 = confusion_matrix(predict_l1_y_test, y_test)
print(confusion_matrix_l1)
accuracy_l1 = accuracy_score(predict_l1_y_test,y_test)
print(accuracy_l1)
```

```python
# L2 Regularlizer - (Ridge)
model_l2 = LogisticRegression(penalty ='l2', solver = 'liblinear', C = 1.0) ## Liblinear id use for identify and ignore the less important feature
model_l2.fit(X_train, y_train)
predict_l2_y_test = model_l2.predict(X_test)
predict_l2_y_test
print(y_test.shape)
print(predict_l2_y_test.shape)

#Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
confusion_matrix_l2 = confusion_matrix(predict_l2_y_test, y_test)
print(confusion_matrix_l2)
accuracy_l2 = accuracy_score(predict_l2_y_test,y_test)
print(accuracy_l2)
```

```python
## Elastic
model_elastic = LogisticRegression(penalty= 'elasticnet', solver = 'saga', C = 1.0, l1_ratio=0.5) ## saga is use for identify and ignore the less important feature
model_elastic.fit(X_train, y_train)
predict_elastic_y_test =model_elastic.predict(X_test)
predict_elastic_y_test
print(y_test.shape)
print(predict_l2_y_test.shape)

#Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
confusion_matrix_elastic = confusion_matrix(predict_elastic_y_test, y_test)
print(confusion_matrix_elastic)
report_elastic = classification_report(predict_elastic_y_test,y_test)
print(report_elastic)
accuracy_elastic = accuracy_score(predict_elastic_y_test,y_test)
print(accuracy_elastic)
```
